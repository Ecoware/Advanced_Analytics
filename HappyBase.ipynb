```python
#import libraries
from google.cloud import bigtable
from google.cloud import happybase
```

```python
#setup client, instance, connection, table
client = bigtable.Client(project='network-sec-analytics', admin=False)
instance = client.instance('nsabigtable')
connection = happybase.Connection(instance=instance)
hbtable = connection.table('ipfix_udp_base_agg') 
```

```python
#read data
for key, data in hbtable.scan(row_start=b'151.140.1.143', row_stop=b'165.130.174.182'):
    print(key, data)
```
